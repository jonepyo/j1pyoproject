## Visualization 


Histogram of Winners by points. San Francisco has a record of having the most points in the championship overall. 

```python
px.histogram(df, x = 'Winner', y = 'Winner Pts', title = 'Winner by Points')
```

Tom Brady have been the most valuable player 4 times which makes a record in NFL championship history. 

```python
px.box(df, x= 'MVP', y = 'Winner Pts', title = 'Winning Team points with MVP')
```

```python
fig3 = px.scatter(df, x = 'Winner Pts', y= 'Loser Pts')
fig3
```

```python
px.pie(MVPCount, 'MVP', width = 800, height = 700)
```

There is only one 4 winning MVP in the Championship. 

```python
df['Difference'] = df['Winner Pts'] - df['Loser Pts']
px.bar(df, x='SB', y = 'Difference', width = 1000, height = 500 )
```

The graph shows that 24th championship had a greatest difference between winner and loser. 

- From the analysis and visualization, I could see that 24th championship has the greatest point difference between winners and losers. Moreover, overall MVP is Tom Brady with 4 championship MVPs.
